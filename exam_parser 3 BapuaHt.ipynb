```python
from lxml import html
import requests

your_link = input('Введите ссылку на нужную страницу в вики:')

page = requests.get(your_link)
tree = html.fromstring(page.content)
links = tree.xpath('//a/@href')

i = 0 
while i < len(links) : 
    print( str(i + 1) + '. ' + str(links[i]) )
    i += 1
```
